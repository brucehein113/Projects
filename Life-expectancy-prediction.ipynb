```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
raw = pd.read_csv('https://raw.githubusercontent.com/myanmards/resource_files/refs/heads/master/regression-one.csv')
data = raw[['Life expectancy', 'Adult Mortality',
       'infant deaths', 'Alcohol', 'percentage expenditure', 'Hepatitis B',
       'Measles', 'BMI', 'under-five deaths', 'Polio', 'Total expenditure',
       'Diphtheria', 'HIV/AIDS', 'GDP', 'Population', 'thinness 1-19 years',
       'thinness 5-9 years', 'Income composition of resources', 'Schooling']].dropna()
```

```python
data.shape
```

```python
data.describe().style.background_gradient('hot')
```

```python
data.corr().style.background_gradient('Greens')
```

```python
life = data[['Life expectancy']]
other = data[['Schooling','Alcohol']]
```

```python
lr.fit(other, life)
```

```python
pred = lr.predict(other)
```

```python
axs = sns.distplot(life, hist = False, label = 'Actual', color = 'blue')
sns.distplot(pred, hist = False, label = 'Predicted', color = 'red', ax = axs)
plt.legend(loc = 'upper left')
plt.title('Effective of Schooling and Alcohol on life')
plt.xlabel('Life expectancy')
plt.ylabel('Schooling and Alcohol density')
plt.show()
```

```python
def classification(x):
    if x <= 9.7:
        return 'Low'
    elif x > 9.7 and x <= 15.2:
        return 'Normal'
    else:
        return 'High'
```

```python
#This is sample dataset
brand = ['Macbook', 'Lenovo', 'Asus', 'Dell', 'Acer','Hp']
man = [1000, 2000, 1500, 1300, 900, 700]
woman = [1600, 700, 1000, 900, 500, 1000]
index = np.arange(len(brand))
bar_width = 0.4
```

```python
fig, ax = plt.subplots()
ax.bar(index,man, bar_width, label='Man')
ax.bar(index + bar_width, woman, bar_width, label='Woman')
ax.set_xticks(index+bar_width)
ax.set_xticklabels(brand)
plt.title('Using Laptop brand')
plt.xlabel('Brands')
plt.ylabel('Units')
plt.legend()
plt.show()
```

```python
plt.barh(brand, man, label = 'Man only')
plt.legend()
plt.title('Laptop brand & Man usage')
plt.xlabel('Usage')
plt.ylabel('Laptop Brand')
plt.show()
```

```python
plt.plot(brand, woman, label='Woman')
plt.plot(brand, man, label='Man')
plt.legend()
plt.title('Laptop Usage')
plt.xlabel('Brands')
plt.ylabel('Units')
plt.show()
```

```python
#OOP Testing

class laptop():
    def __init__(self, brand, color, cpu, ram, storage):
        self.brand = brand
        self.color = color
        self.cpu = cpu
        self.ram = ram
        self.storage = storage

    def get_brand(self):
        print(f'Laptop Brand is ', self.brand)

    def get_color(self):
        print(f'Laptop Color is ', self.color)

    def get_cpu(self):
        print(f'Laptop Processor is ', self.cpu)

    def get_ram(self):
        print(f'Laptop Memory is ', self.ram)

    def get_storage(self):
        print(f'Laptop Storage is ', self.storage)

    def get_specs(self):
        print(f'Laptop Specifications are ', self.brand, self.color, self.cpu, self.ram, self.storage)
```
